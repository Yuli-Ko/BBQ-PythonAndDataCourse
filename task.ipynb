```python
input_list = [2, 5, 8, 11, 14, 17, 20]

result_list = []

for number in input_list:
    if number % 2 and number > 5:
        result_list.append(number**2)

print(result_list)
```

```python
list1 = [1, 2, 3, 4]
list2 = [5, 6, 7, 8]

sum_list = []
for elem1, elem2 in zip(list1, list2):
    sum_list.append(elem1 + elem2)

print(sum_list)
```

```python
nested_list = [['Apfel', 'Banane'], ['Banane', 'Erdbeere', 'Kiwi'], ['Orange', 'Apfel']]
result_set = set()

for all in nested_list:
    for word in all:
        result_set.add(word)

result_list = list(result_set)

print(result_list)
```

```python
word_list = ['Apfel', 'Banane', 'Erdbeere', 'Kiwi', 'Orange']

sorted_list = sorted(word_list)

filtered_list = []
for word in sorted_list:
    if word.startswith('B') or len(word) <= 4:
        filtered_list.append(word)

print(filtered_list)
```
