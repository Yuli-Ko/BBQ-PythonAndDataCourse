```python


# Группировка данных по полу и состоянию здоровья и подсчет количества в каждой группе
grouped_health_condition = data.groupby(['Gender','age group', 'Medical Condition']).size().unstack()

# Создание столбчатой гистограммы для каждой комбинации группы и пола
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(12, 10))

genders = ['женщин', 'мужчин']

for i, (gender_index, gender_group) in enumerate(grouped_health_condition.groupby('Gender')):
    gender = genders[gender_index]
    gender_group.unstack().plot(kind='bar', ax=ax[i], color=['skyblue', 'salmon', 'lightgreen', 'orange', 'purple'])
    ax[i].set_title('Распределение состояния здоровья у ' + gender)
    ax[i].set_xlabel('Состояние здоровья')
    ax[i].set_ylabel('Частота')
    ax[i].tick_params(axis='x', rotation=45)
    ax[i].grid(axis='y', linestyle='--')

plt.tight_layout()
plt.show()
```

```python

# Set Seaborn style
sns.set(style="whitegrid")

# Create a bar plot for gender distribution
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='Gender')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Create a bar plot for blood type distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Blood Type')
plt.title('Blood Type Distribution')
plt.xlabel('Blood Type')
plt.ylabel('Count')
plt.show()

# Create a bar plot for the frequency of medical conditions
plt.figure(figsize=(12, 6))
sns.countplot(data=data, y='Medical Condition')
plt.title('Frequency of Medical Conditions')
plt.xlabel('Count')
plt.ylabel('Medical Condition')
plt.show()

# Create a histogram for age distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='Age', bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Create a bar plot for the frequency of insurance providers
plt.figure(figsize=(10, 6))
sns.countplot(data=data, y='Insurance Provider')
plt.title('Frequency of Insurance Providers')
plt.xlabel('Count')
plt.ylabel('Insurance Provider')
plt.show()

# Create a box plot for billing amount distribution
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, y='Billing Amount')
plt.title('Billing Amount Distribution')
plt.ylabel('Billing Amount')
plt.show()

# Create a bar plot for admission types
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='Admission Type')
plt.title('Admission Type Distribution')
plt.xlabel('Admission Type')
plt.ylabel('Count')
plt.show()

# Create a bar plot for the frequency of medication
plt.figure(figsize=(12, 6))
sns.countplot(data=data, y='Medication')
plt.title('Frequency of Medication')
plt.xlabel('Count')
plt.ylabel('Medication')
plt.show()

# Create a pie chart for test results distribution
plt.figure(figsize=(8, 8))
data['Test Results'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Test Results Distribution')
plt.ylabel('')
plt.show()
```

```python

# Define bins and labels for age groups
bins = [0, 20, 35, 60, 75, 92, float('inf')]
labels = ['rare', 'Medium rare', 'Medium', 'Medium well', 'Well done', 'Burned']

# Create a new column for age groups
data['Age Group'] = pd.cut(data['Age'], bins=bins, labels=labels, right=False)

# Create a pivot table to count the occurrences of medical conditions by age group and gender
pivot_table = data.pivot_table(index='Medical Condition', columns=['Age Group', 'Gender'], aggfunc='size', fill_value=0)

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap='viridis', annot=True, fmt='d', linewidths=.5)
plt.title('Medical Condition Distribution by Age Group and Gender')
plt.xlabel('Age Group')
plt.ylabel('Medical Condition')
plt.show()
```
