```python
import requests
from requests.models import PreparedRequest
import time
import json

PARAM_OFFSET = 0
PARAM_LIMIT = 40
PARAM_CATEGORY_ID = 1602
PARAM_REGION_ID = 9
PARAM_CITY_ID = 62
PARAM_CURRENCY = "USD"
PARAM_MAX_PRICE = 56000

FETCH_INTERVAL_SECONDS = 30

def scanDomiciles():
    #response = requests.get("https://www.olx.ua/api/v1/offers/?offset=0&limit=40&category_id=1602&region_id=9&city_id=62&currency=USD&filter_float_price%3Ato=56000")
    data = []
    maxElements = 9999999
    startIndex = PARAM_OFFSET
    while maxElements > startIndex:
        response = fetchData(startIndex, PARAM_LIMIT, PARAM_CATEGORY_ID, PARAM_REGION_ID, PARAM_CITY_ID, PARAM_CURRENCY, PARAM_MAX_PRICE)
        startIndex = startIndex + PARAM_LIMIT

        print("Max elements: " + str(maxElements))
        print("Offset: " +str(startIndex))

        dataJson = response.json()

        print(dataJson["data"][0]["params"][0]["value"]["value"])

        maxElements = dataJson["metadata"]["total_elements"]
        domiciles = dataJson["data"]
        data.extend(domiciles)
        if "next" not in dataJson["links"]:
            break
        nextLink = dataJson["links"]["next"]["href"]
        #print("------------------------------")
        #print(domiciles)
        print("Fetching next link: " + nextLink)
        time.sleep(FETCH_INTERVAL_SECONDS)
    print("Scan complete")
    print("Total number of elements: " + str(len(data)))
    with open("/tmp/data.json", "w") as f:
        json.dump({"domiciles": data}, f)

def fetchData(offset, limit, category_id, region_id, city_id, currency, max_price):
    return requests.get("https://www.olx.ua/api/v1/offers/", params={"offset": offset, "limit": limit, "category_id": category_id, "region_id": region_id, "city_id": city_id, "currency": currency, "filter_float_price:to": max_price})


if __name__ == '__main__':
    scanDomiciles();
```
